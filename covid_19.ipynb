```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_countries(countries, df):
    plt.figure(figsize=(12,6))

    for country_name in countries:
        # Select rows for the country
        country_df = df.loc[country_name]

        # If only one row (Series)
        if isinstance(country_df, pd.Series):
            country_df = pd.to_numeric(country_df, errors='coerce')
            country_df.dropna().plot(label=country_name)

        # If multiple rows (DataFrame with provinces)
        else:
            country_df = country_df.drop(['Province/State'], axis=1, errors='ignore')
            country_df = country_df.apply(pd.to_numeric, errors='coerce')
            country_df = country_df.sum(axis=0)
            country_df.plot(label=country_name)

    plt.title("COVID Trend Comparison")
    plt.xlabel("Date")
    plt.ylabel("Cases")
    plt.legend()
    plt.show()
```

```python


plot_countries(["China", "India", "Spain"], corona_dataset_aggregrate)
```

```python
#calcualate a good measures
# Select China row(s)
china_df = corona_dataset_aggregrate.loc["China"]

# Drop non-numeric columns and convert the rest to numbers
china_df_numeric = china_df.apply(pd.to_numeric, errors='coerce')

# Drop NaNs just in case
china_df_numeric = china_df_numeric.dropna()

# Plot
china_df_numeric.plot(title="COVID Trend in China", figsize=(10,5))
plt.ylabel("Cases")
plt.xlabel("Date")
plt.show()
```

```python
#calculate the first derivate of the curve
# Select China row(s)
china_df = corona_dataset_aggregrate.loc["China"]

# If multiple provinces, sum first
if isinstance(china_df, pd.DataFrame):
    china_df = china_df.drop(['Province/State'], axis=1, errors='ignore')
    china_df = china_df.apply(pd.to_numeric, errors='coerce')
    china_df = china_df.sum(axis=0)
else:
    china_df = china_df.apply(pd.to_numeric, errors='coerce')

# Calculate daily new cases
china_daily = china_df.diff()

# Plot daily new cases
china_daily.plot(title="Daily New COVID Cases in China", figsize=(10,5))
plt.ylabel("Daily Cases")
plt.xlabel("Date")
plt.show()
```

```python
countries = list(corona_dataset_aggregrate.index)
max_infection_rates = []

for c in countries:
    country_df = corona_dataset_aggregrate.loc[c]

    # If multiple provinces, sum them
    if isinstance(country_df, pd.DataFrame):
        country_df = country_df.drop(['Province/State'], axis=1, errors='ignore')
        country_df = country_df.apply(pd.to_numeric, errors='coerce')
        country_df = country_df.sum(axis=0)
    else:
        country_df = country_df.apply(pd.to_numeric, errors='coerce')

    # Daily new cases
    daily_cases = country_df.diff()
    max_infection_rates.append(daily_cases.max())

# Add new column
corona_dataset_aggregrate['max_infection_rate'] = max_infection_rates

# Check results
corona_dataset_aggregrate
```

```python
#create a new dataframe
corona_data = pd.DataFrame(corona_dataset_aggregrate["max_infection_rate"])
corona_data
```
